```python
df_filtered = df_filtered.dropna(subset=['outname'])

# Convert categorical variables to numerical values if needed
# For simplicity, assume 'outname' is the target and all other columns are features
X = df_filtered.drop(columns=['outname', 'group'])
y = df_filtered['outname']
X = pd.get_dummies(X, drop_first=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import AdaBoostClassifier, StackingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.neural_network import MLPClassifier

df_test['outcome'] = df_test['outcome'].astype('category').cat.codes
df_test['outname'] = df_test['outname'].astype('category').cat.codes

X = df_test[['BSAE', 'low_sae', 'up_sae', 'outcome']]
y = df_test['outname']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

ada_clf = AdaBoostClassifier(n_estimators=50, random_state=42)
ada_clf.fit(X_train, y_train)
y_pred_ada = ada_clf.predict(X_test)
print("AdaBoost Classification Report:")
print(classification_report(y_test, y_pred_ada, zero_division=0))

base_classifiers = [
    ('decision_tree', DecisionTreeClassifier()),
    ('logistic_regression', LogisticRegression(max_iter=1000))
]
stacking_clf = StackingClassifier(estimators=base_classifiers, final_estimator=LogisticRegression())

stacking_clf.fit(X_train, y_train)
y_pred_stack = stacking_clf.predict(X_test)
print("Stacking Classification Report:")
print(classification_report(y_test, y_pred_stack, zero_division=0))

mlp_clf = MLPClassifier(hidden_layer_sizes=(100, 50), max_iter=1000, random_state=42)
mlp_clf.fit(X_train, y_train)
y_pred_mlp = mlp_clf.predict(X_test)

print("Neural Network Classification Report:")
print(classification_report(y_test, y_pred_mlp, zero_division=0))

```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

conf_matrix_ada = confusion_matrix(y_test, y_pred_ada)
plt.figure(figsize=(10, 7))
sns.heatmap(conf_matrix_ada, annot=True, fmt='d', cmap='Blues', 
            xticklabels=['Class 0', 'Class 1'], yticklabels=['Class 0', 'Class 1'])
plt.title('Confusion Matrix - AdaBoost')
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.show()

conf_matrix_stack = confusion_matrix(y_test, y_pred_stack)
plt.figure(figsize=(10, 7))
sns.heatmap(conf_matrix_stack, annot=True, fmt='d', cmap='Blues', 
            xticklabels=['Class 0', 'Class 1'], yticklabels=['Class 0', 'Class 1'])
plt.title('Confusion Matrix - Stacking Classifier')
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.show()

```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.neural_network import MLPClassifier
import pandas as pd

# Assuming you have already trained the model and made predictions
mlp_clf = MLPClassifier(hidden_layer_sizes=(100, 50), max_iter=1000, random_state=42)
mlp_clf.fit(X_train, y_train)
y_pred_mlp = mlp_clf.predict(X_test)

# Compute the confusion matrix
cm = confusion_matrix(y_test, y_pred_mlp)
class_names = np.unique(y_test)  # Or define your class names

# Display the confusion matrix
plt.figure(figsize=(10, 7))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_names, yticklabels=class_names)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()

# Distribution of Predictions
plt.figure(figsize=(10, 6))
sns.countplot(x=y_pred_mlp, palette='viridis')
plt.title('Distribution of Predictions by MLP Classifier')
plt.xlabel('Predicted Class')
plt.ylabel('Count')
plt.xticks(ticks=np.arange(len(class_names)), labels=class_names)
plt.show()

# Classification Report
report = classification_report(y_test, y_pred_mlp, target_names=class_names, output_dict=True)
report_df = pd.DataFrame(report).transpose()

print("Classification Report:")
print(report_df)

# Optional: Plot a heatmap of classification report metrics
plt.figure(figsize=(12, 6))
sns.heatmap(report_df.iloc[:-1, :-1], annot=True, cmap='Blues', fmt='.2f')
plt.title('Classification Report Metrics Heatmap')
plt.xlabel('Metrics')
plt.ylabel('Classes')
plt.show()
```

```python
## Conclusion

In this analysis, we explored various classification models to predict addiction-related outcomes using the NASDAH dataset, which provides comprehensive information on substance use and mental health over several years. Our goal was to determine which individuals are at risk of addiction and identify the most impactful features for classification. The models evaluated included AdaBoost, Stacking Classifier, Neural Network (MLP), Bagging Classifier, and Pasting Classifier.

**Key Findings:**

1. **Model Performance:**
   - **AdaBoost Classifier:** Achieved high accuracy but struggled with precision and recall for certain classes, indicating some imbalance in predictions.
   - **Stacking Classifier:** Provided a balanced performance with a good overall accuracy, but similar to AdaBoost, it showed varying precision and recall across classes.
   - **Neural Network (MLP):** Offered a nuanced view with varying performance metrics. The neural network was sensitive to the feature scaling and the complexity of the data.
   - **Bagging Classifier:** Demonstrated robust performance with better stability across different classes, mitigating some of the model biases observed in AdaBoost and Stacking.
   - **Pasting Classifier:** Provided similar benefits as Bagging, with emphasis on sample subsets rather than bootstrap sampling, showing competitive results in classification accuracy.

2. **Feature Importance:**
   - Important features such as `BSAE`, `low_sae`, `up_sae`, and `ste_sae` were critical in determining addiction-related outcomes. These features contributed significantly to the classification models' performance.

3. **Challenges and Limitations:**
   - Despite high overall accuracy, precision and recall metrics varied significantly across classes, highlighting challenges in predicting less prevalent classes.
   - The presence of undefined metrics and warnings suggests that further tuning and data preprocessing might be necessary to enhance model performance.

**Future Directions:**

Given the promising results with the NASDAH dataset, the next step involves applying these predictive models to similar datasets, such as those from India, to assess the transferability of our findings. Adapting our models to the Indian context will require careful consideration of regional data characteristics and potentially recalibrating models to account for differences in substance use patterns and mental health indicators.

By leveraging these models and refining them for diverse datasets, we can enhance our understanding of addiction patterns globally and develop targeted interventions to support affected individuals. The insights gained from this study not only contribute to predictive analytics in addiction research but also pave the way for more effective strategies in public health across different regions.
```
